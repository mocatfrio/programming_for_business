# what is an array

# Creating an array
#### An array is a special variable, which can hold more than one value at a time.
#### Note: Python does not have built-in support for Arrays, but Python Lists can be used instead.

```python
numbers=[1,2,3,4,5]
```

# Access the Elements of an Array
#### you can access an array by using the index of this element note that index of the element start from 0 this mean that index of the element in an array equal the ranking of this element -1

```python
print(numbers[0]) #print the first element 
print(numbers[:]) #print the whole array
print(numbers[-1]) #print the last element in the array
```

#### you can read the whole array elements using for loop

```python
odd=[1,3,5,9]
for i in odd:
    print(i,end=" ")
```

# Inserting Items Into array

#### Inserting a new element into an Array can take many forms:

#### 1-Inserting a new element at the end of the Array.
#### 2-Inserting a new element at the beginning of the Array.
#### 3-Inserting a new element at any given index inside the Array.


### 1-Inserting at the End of an Array
#### At any point in time, we know the index of the last element of the Array, as we've kept track of it in
#### our length variable. All we need to do for inserting an element at the end is to assign the new 
#### element to one index past the current last element.


#### in python you can do this operation use the build in function append()

```python
arr=[1,2,3,4,5]
print(len(arr)) #len() is a build in function used for printing the length of array
arr.append(7)
print(arr)
print(len(arr)) #notice the changing in the length
```

### Inserting at the Start of an Array
#### To insert an element at the start of an Array, we'll need to shift all other elements in the Array to the right by one index to create space for the new element. This is a very costly operation, since each of the existing elements has to be shifted one step to the right. The need to shift everything implies that this is not a constant time operation. In fact, the time taken for insertion at the beginning of an Array will be proportional to the length of the Array. In terms of time complexity analysis, this is a linear time complexity: O(N)O(N), where NN is the length of the Array.

#### you can use the build in function insert()to insert an item in the beginning of array
#### insert take two arrguments the first one is the index of the item you want to insert ,the second one is the value of the item you want to insert

```python
a=[2,3,4]
a.insert(0,1) #notice that we want to insert 1 in the bgainning of the array a so we give it the inex 0
print(a)
```

### Inserting Anywhere in the Array
#### Similarly, for inserting at any given index, we first need to shift all the elements from that index onwards one position to the right. Once the space is created for the new element, we proceed with the insertion. If you think about it, insertion at the beginning is basically a special case of inserting an element at a given indexâ€”in that case, the given index was 0.

#### again we can use the insert() function by changing the index to insert in anywhere we want

```python
arr=[1,2,3,5]
arr.insert(3,4)
print(arr)
```
